```python
# Convert date column to datetime
weather['date'] = pd.to_datetime(weather['date'])
```

```python
# Merge datasets
df = pd.merge(weather, locations, on='location_id')
```

```python
# checking the number of missing values
print(df.isnull().sum())
```

```python
# Handling Time-Based Features (Sunrise/Sunset) - Drop
df = df.drop(columns=['sunrise', 'sunset'])
```

```python
# Handling Numerical Weather Features
# For wind, radiation, and precipitation values, will be used in advanced imputation:

# Create groups for district and month
group_keys = ['location_id', df['date'].dt.month]

# Define columns needing imputation
num_cols = [
    'precipitation_hours (h)',
    'wind_speed_10m_max (km/h)',
    'wind_gusts_10m_max (km/h)',
    'wind_direction_10m_dominant (°)',
    'shortwave_radiation_sum (MJ/m²)',
    'et0_fao_evapotranspiration (mm)'
]

# Fill missing values with district-month averages
for col in num_cols:
    df[col] = df.groupby(group_keys)[col].transform(
        lambda x: x.fillna(x.mean()))
```

```python
# Verify no missing values remain
print(df.isnull().sum())
```
